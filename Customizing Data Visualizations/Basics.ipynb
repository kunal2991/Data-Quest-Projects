```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
recent_grads = pd.read_csv('recent-grads.csv')
print(recent_grads.head())
print(recent_grads.tail())
recent_grads.describe()
```

```python
print(recent_grads.shape[0])
recent_grads_dropna = recent_grads.dropna()
print(recent_grads_dropna.shape[0])
```

```python
fig = plt.figure(figsize=(8,8))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
ax1.hist(recent_grads_dropna['ShareWomen'])
ax2.scatter(recent_grads_dropna['Unemployment_rate'],recent_grads_dropna['ShareWomen'])
ax3.scatter(recent_grads_dropna['ShareWomen'],recent_grads_dropna['Unemployment_rate'])
ax4.hist(recent_grads_dropna['Unemployment_rate'])
ax1.get_xaxis().set_visible(False)
ax2.get_xaxis().set_visible(False)
ax1.set_ylabel('ShareWomen')
ax3.set_xlabel('ShareWomen')
ax3.set_ylabel('Unemployment_rate')
ax4.set_xlabel('Unemployment_rate')
fig.subplots_adjust(wspace=0,hspace=0)
ax1.set_xlim(0,30)
ax2.set_xlim(0.0,1.0)
ax3.set_xlim(0.0,1.0)
ax3.set_ylim(0.0,0.20)
ax4.set_xlim(0.0,0.20)
ax1.set_yticklabels([0,5,10,15,20,25,30])
ax3.set_yticklabels([0.00,0.05,0.10,0.15])
ax3.set_xticklabels([0.0,0.2,0.4,0.6,0.8],rotation=90)
ax4.set_xticklabels([0.00,0.05,0.10,0.15,0.20],rotation=90)
plt.show()
```

```python
ShareMen = recent_grads['Men'] / recent_grads['Total']
print(ShareMen)
recent_grads['ShareMen'] = ShareMen
```

```python
Arts_Categ = recent_grads[recent_grads['Major_category']== "Arts"]
#Arts_Categ.set_index('Major',inplace=True)
print(Arts_Categ)
```

```python
import numpy as np
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
locs = np.arange(0,8)
#print(locs)
Arts_Categ_list = Arts_Categ['ShareMen'].tolist()
#print(recent_grads_list)
bar_1 = plt.bar(locs,Arts_Categ_list,width=0.35)
major_list = Arts_Categ['Major'].tolist()
print(major_list)
ax.set_xticklabels(major_list,rotation=90)
offset_locs = locs + 0.35
#print(offset_locs)
bar_2 = plt.bar(offset_locs,Arts_Categ['ShareWomen'].tolist(),width=0.35,color='green')
ax.set_xticks(offset_locs)
plt.legend((bar_1,bar_2),('ShareMen','ShareWomen'),loc='upper left')
ax.grid()
plt.show()
```
