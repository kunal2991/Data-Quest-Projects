```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
pixar_movies = pd.read_csv("PixarMovies.csv")
dimensions = pixar_movies.shape
print(dimensions)
print(pixar_movies.head)
```

```python
print(pixar_movies.dtypes)
```

```python
print(pixar_movies.describe())
```

```python
domestic_stripped = pixar_movies['Domestic %'].str.rstrip('%').astype('float')
print(domestic_stripped)
international_stripped = pixar_movies['International %'].str.rstrip('%').astype('float')
print(international_stripped)
pixar_movies['Domestic %'] = domestic_stripped
pixar_movies['International %'] = international_stripped
print(pixar_movies)
```

```python
pixar_movies['IMDB Score'] = pixar_movies['IMDB Score']*10
print(pixar_movies)
```

```python
filtered_pixar = pixar_movies.loc[0:13]
print(filtered_pixar)
```

```python
pixar_movies.set_index('Movie',inplace=True)
filtered_pixar.set_index('Movie',inplace=True)
print(pixar_movies.head())
```

```python
critics_review = pixar_movies[['RT Score','IMDB Score','Metacritic Score']]
critics_review.plot()
plt.show()
```

```python
critics_review.plot(figsize=(10,6))
plt.show()
```

```python
critics_review.plot(kind='box',figsize=(9,5))
plt.show()
```

```python
revenue_proportions = filtered_pixar[['Domestic %','International %']]
revenue_proportions.plot(kind='bar',stacked=True)
plt.show()
```
