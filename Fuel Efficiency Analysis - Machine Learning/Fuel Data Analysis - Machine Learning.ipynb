```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

mpg = pd.read_csv('auto-mpg.data.txt',delim_whitespace=True)
print(mpg.head(3))
```

Importing data. Notice that there are no column names, so let's set the column names using pandas columns attribute.

```python
columns = ["mpg", "cylinders", "displacement", "horsepower", "weight", "acceleration", "model year", "origin", "car name"]
cars = pd.read_table("auto-mpg.data.txt", delim_whitespace=True, names=columns)
print(cars.head(5))
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
cars.plot("weight", "mpg", kind='scatter', ax=ax1)
cars.plot("acceleration", "mpg", kind='scatter', ax=ax2)
plt.show()
```

From the basic scatter plot analysis between (weight vs mpg) and (acceleration vs mpg) we can observe that there is a general negative linear correlation between weight and mpg indicating that the efficiency or miles per gallon decreases with an increase in the weight, which is fairly obvious.

Also, there is no clear trend or a linear relationship between acceleration and mpg.

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(cars[['weight']],cars[['mpg']])
```

In the above code cell, we have initialized a Linear Regressor and are fitting a model on the available data to help predict values on new data that could occur.
Linear Regression or any linear model follows the following equation: y = mx + c, where y is the dependent var, m is slope, x is the independent var and c is the 'y-intercept' of the line on the X-Y coordinate plane.

```python
predictions = lr.predict(cars[['weight']])
print(predictions[0:5])
print(cars['mpg'][0:5])
```

Once the model has been trained, its now time to predict values using the 'predict' function on the regression model. We are using the trainind data as our test data (which is not generally the practice) to understand if there is any underfitting or not - Underfitting: When we train our data using data X, and for the same data X, our model cannot predict properly or accurately.

```python
plt.scatter(cars['weight'],cars['mpg'],c='red')
plt.scatter(cars['weight'],predictions,c='blue')
plt.show()
```

As we can now see from the scatter plot above, the BLUE line is our Model predictions using the RED data. The BLUE line indicates a best fit linear regression line to predict 'mpg' values given 'weight' values.

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(cars['mpg'],predictions)
print(mse)
```

The next step is to determine the 'Error Score' or some sort of an error metric for our predictive model to understand its effectiveness.
A common error metric that we use is the Mean Squared Error or MSE. It is essentially the sum of squared differences between the predicted values and the actual data values.
Here, RMSE or Root Mean Squared Error is better fitting as an error metric since we have MSE in terms of Miles Per Gallon Square instead of MPG.

```python
rmse = mse ** (1/2)
print(rmse)
```
