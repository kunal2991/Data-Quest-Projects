```python
import requests
import json

headers = {"Authorization": "token Enter your token"}

response = requests.get("https://api.github.com/users/kunal2991", headers=headers)
data = response.json()
print(data)
```

Most APIs have a security feature i.e authentication feature. For this, first you have to generate a token and then use that token to access the API.
We use an Authorization header to access data.
Note: Need to use keyword 'token'.

```python
response = requests.get('https://api.github.com/users/torvalds', headers=headers)
data = response.json()
print(data)
```

Using the token, we can retrieve information about other users as well as you can see above, we have accessed data about Linus Torvalds.

```python
response = requests.get('https://api.github.com/users/VikParuchuri/orgs', headers=headers)
data = response.json()
print(data)
```

```python
response = requests.get('https://api.github.com/repos/octocat/Hello-World', headers=headers)
hello_world = response.json()
print(hello_world)
```

We can use the Github API to access information of users as well as organizations. The above example, we are accessing information of a user 'octocat' within the 'DataQuest' Github organization.

```python
#Pagination
params = {"per_page": 50, "page": 2}
response = requests.get('https://api.github.com/users/VikParuchuri/starred', params=params)
page2_repos = response.json()
print(page2_repos)
```

In order to avoid a huge amount of data being retrieved which in turn will overload the server, the API provider will only return a certain number of records per page. 
We can use the 'params' header to access or modify the parameters like we have done above by restricting the number of records per page to 50 and selecting the 2nd page.

```python
response = requests.get('https://api.github.com/user', headers=headers)
user = response.json()
print(user)
```

The above code is for a User-Level Endpoint and will return the data about the user with which we have authorized.

```python
payload = {'name':'learning-about-apis'}
response = requests.post("https://api.github.com/user/repos", json=payload, headers=headers)
status = response.status_code
print(status)
```

Just like GET requests, we have POST requests which are used to send information to server and create objects on the server.
In the above cell, we are creating a repo called 'learning-about-apis' on the currently authenticated user's GitHub repo.

```python
payload = {'description':'Learning about requests!','name':'learning-about-apis'}
response = requests.patch('https://api.github.com/user/repos/learning-about-apis',json=payload, headers=headers)
status = response.status_code
print(status)
```

We also have PUT and PATCH requests. PATCH requests are used to update certain attributes of an object instead of replacing the entire object.
PUT requests are used to replace the current version of an object with the one that we want, essentially replacing the entire object.

```python
response = requests.delete('https://api.github.com/user/repos/learning-about-apis',headers=headers)
print(response.status_code)
```
